```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
from matplotlib import style
from sklearn.datasets import load_iris 
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns= iris.feature_names)
df['target'] = iris.target
df.head()
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
```

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'],color="green",marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'],color="blue",marker='.')
plt.scatter(df2['sepal length (cm)'], df1['sepal width (cm)'],color="red",marker='*')
plt.title("Sepal Length and Width of different Iris flowers")
```

```python
unique_targets = np.unique(df['target'])
unique_targets
```

```python
# Average values for target 0 
avg0_sepal_length = np.average(df0['sepal length (cm)'])
avg0_sepal_width = np.average(df0['sepal width (cm)'])
avg0_petal_length = np.average(df0['petal length (cm)'])
avg0_petal_width = np.average(df0['petal width (cm)'])

# Average values for target 1 
avg1_sepal_length = np.average(df1['sepal length (cm)'])
avg1_sepal_width = np.average(df1['sepal width (cm)'])
avg1_petal_length = np.average(df1['petal length (cm)'])
avg1_petal_width = np.average(df1['petal width (cm)'])

# Average values for target 2 
avg2_sepal_length = np.average(df2['sepal length (cm)'])
avg2_sepal_width = np.average(df2['sepal width (cm)'])
avg2_petal_length = np.average(df2['petal length (cm)'])
avg2_petal_width = np.average(df2['petal width (cm)'])
```

```python
sepal_lengths = (avg0_sepal_length, avg1_sepal_length, avg2_sepal_length)
sepal_widths = (avg0_sepal_width, avg1_sepal_width, avg2_sepal_width)
petal_lengths = (avg0_petal_length, avg1_petal_length, avg2_petal_length) 
petal_widths = (avg0_petal_width, avg1_petal_length, avg2_petal_width)
```

```python
plt.figure(figsize=(14,7)) 
plt.bar(unique_targets, sepal_lengths, color='royalblue', width = 0.2, label = 'average sepal length')
plt.bar(unique_targets + 0.2, sepal_widths, color='orange', width = 0.2, label = 'average sepal width')
plt.bar(unique_targets + 0.4, petal_lengths, color='green', width = 0.2, label = 'average petal length')
plt.bar(unique_targets + 0.6, petal_widths, color='red', width = 0.2, label = 'average petal width')
plt.xticks(unique_targets + 0.3, iris.target_names)
plt.xlabel("Iris Type") 
plt.ylabel("measurement in cm")
plt.legend()
plt.title("Average features of different Iris flowers")
```

```python
df['flower_name'] =df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='flower_name', y='sepal length (cm)', data=df)

# Add labels and title
plt.xlabel('Target')
plt.ylabel('Sepal Length (cm)')
plt.title('Boxplot of Sepal Length by Target')
```
